## Maps

```python
moyenne = reviews.ActiveCases.mean()
moyenne
```

```python
reviews.ActiveCases.map(lambda cas_actif: 1 if cas_actif > moyenne else 0)
# mettre a 1 si le 'ActiveCases' > moyenne des ActiveCases , 0 sinon
```

```python
moyenne_TotalDeaths = reviews.TotalDeaths.mean()
moyenne_TotalDeaths
```

```python
#mettre a 1 si TotalDeaths > moyenne_TotalDeaths 0 sinon
def remean_points(row):
    row.TotalDeaths = 1 if row.TotalDeaths > moyenne_TotalDeaths else 0
    return row

reviews.apply(remean_points, axis='columns')  # axis='columns'??? 
```

# 4 Grouping and Sorting

```python
# reviews = pd.read_csv(csv_file_name, index_col=0)
pd.set_option("display.max_rows", 9) 
```

```python
# reviews
```

### Groupwise analysis

```python
reviews.groupby('Continent').Continent.count()
```

```python
reviews.groupby('Continent').Population.min()
```

```python
reviews.groupby(['Country/Region', 'Continent']).apply(lambda df: df.loc[df.Population.idxmin()])
```

```python
reviews.groupby(['Continent']).Population.agg([len, min, max])
```

### Multi-indexes

```python
countries_reviewed = reviews.groupby(['Continent', 'Country/Region']) .Population.agg([max])
countries_reviewed
# print(dir(countries_reviewed))
```

```python
mi = countries_reviewed.index
type(mi)
```

```python
countries_reviewed.reset_index()
```

### Sorting

```python
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='max')
```

```python
countries_reviewed.sort_values(by='max', ascending=False)
```

```python
countries_reviewed.sort_index()
```

```python
countries_reviewed.sort_values(by=['Continent', 'max'])
```
